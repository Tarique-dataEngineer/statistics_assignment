```python
from datetime import datetime
```

```python
#function to convert into datetime format
def convert_to_date(date_str):

        # Check if the string has at least 4 characters (YYYY)
    if len(date_str) >= 4:
        date_obj = datetime.strptime(date_str[:4], "%Y")
            
            # If there are additional characters (DDD), add them to the date
        if len(date_str) > 4:
            day_of_year = int(date_str[4:])
            date_obj = date_obj.replace(day=1, month=1) + pd.DateOffset(days=day_of_year - 1)
                
        return date_obj.strftime("%Y-%m-%d")
```

```python
df['Debut']=df['Debut'].apply(convert_to_date)
```

4) Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

```python
df.head()
```

```python
data1=df[['Team','GoalsFor','GoalsAgainst']]
```

```python
data1.info()
```

```python
data1[['GoalsFor','GoalsAgainst']]=data1[['GoalsFor','GoalsAgainst']].astype(str).astype(int)
```

```python
data1.info()
```

```python
# Define Goal_diff_count fun()
def Goal_diff_count(dataframe):
    # Calculating goal difference and add it as a new column
    dataframe['GoalDifference'] = dataframe['GoalsFor'] - dataframe['GoalsAgainst']
    
    # Find the team with maximum goal difference
    max_goal_diff_team = dataframe[dataframe['GoalDifference'] == dataframe['GoalDifference'].max()]
    
    # Find the team with minimum goal difference
    min_goal_diff_team = dataframe[dataframe['GoalDifference'] == dataframe['GoalDifference'].min()]
    
    return dataframe, max_goal_diff_team, min_goal_diff_team
```

```python
# Call the Goal_diff_count function
df1, max_goal_diff_team, min_goal_diff_team = Goal_diff_count(data1)
```

```python
#Team with maximum goal difference
max_goal_diff_team
```

```python
#Team with minimum goal difference
min_goal_diff_team
```
